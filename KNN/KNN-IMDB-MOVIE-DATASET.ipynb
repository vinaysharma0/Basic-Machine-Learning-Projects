```python
def recommend_movies(movie_query, k_recommendations):
    raw_movie_data = []
    with open('IMDB-cleaned-DATASET.csv','r') as md:
        next(md) #discard the first line
        
        for line in md.readlines():
            data_row = line.strip().replace('"',"").split(',')
            raw_movie_data.append(data_row)
    print('raw_movie_data : ',raw_movie_data[:5])
    
    movies_rec_data = []
    for row in raw_movie_data:
        data_row = list(map(float, row[2:]))
        movies_rec_data.append(data_row)
        
    print('movies_rec_data : ',movies_rec_data[:5])
    
    rec_indices, _ = KNN.knn(movies_rec_data, movie_query, k = k_recommendations, dist_fn = KNN.euclidean_dist, choice_fn = lambda x: None)
    print('rec_indices : ',rec_indices[:5])
    movie_recommendations = []
    for _, index in rec_indices:
        movie_recommendations.append(raw_movie_data[index])
    return movie_recommendations
```

```python
LInferno = data.iloc[3,2:].values
LInferno
```

```python
recommended_movies = recommend_movies(LInferno, 5)
```
